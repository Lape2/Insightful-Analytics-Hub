```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv')
```

```python
# Display basic information about the dataset
print("Dataset Information:")
print(df.info())

# Display the first few rows of the dataset
print("\nFirst Few Rows of the Dataset:")
print(df.head())
```

```python
# Summary statistics of numerical columns
print("\nSummary Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Drop columns not needed for analysis
# List of columns to drop
columns_to_drop = ['id', 'dateAdded', 'dateUpdated', 'asins', 'imageURLs', 'keys', 'manufacturer', 
                   'manufacturerNumber', 'reviews.date', 'reviews.dateSeen', 'reviews.didPurchase', 
                   'reviews.doRecommend', 'reviews.id', 'reviews.numHelpful', 'sourceURLs', 'primaryCategories', 
                  'reviews.sourceURLs', 'reviews.username']

# Drop the columns
data = df.drop(columns=columns_to_drop)

# Display the first few rows of the cleaned dataset
print(data.head())
```

```python
# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Visualize the distribution of review ratings
plt.figure(figsize=(8, 6))
sns.countplot(x='reviews.rating', data=data)
plt.title('Distribution of Review Ratings')
plt.xlabel('Reviews Rating')
plt.ylabel('Count')
plt.show()

```

```python
# Explore the distribution of review text lengths
data['reviews.length'] = data['reviews.text'].apply(len)
plt.figure(figsize=(8, 6))
sns.histplot(data['reviews.length'], bins=50)
plt.title('Distribution of Review Text Lengths')
plt.xlabel('Review Text Length')
plt.ylabel('Count')
plt.show()
```

**Findings:**

Upon visualizing the review ratings, it's evident that the distribution skews to the right. This indicates that a larger proportion of the dataset comprises high review ratings, particularly rating 5. Conversely, when examining the distribution of review text length, it skews to the left. The majority of review text lengths fall within the range of 0 to 1800 characters.

**Identify Most Reviewed Amazon Products:**

Summary: Count the number of reviews for each product to identify the most reviewed ones.

Task: Calculate the review count for each product and rank them based on review count.

```python
# Calculate review count for each product
review_count_per_product = data.groupby('name')['reviews.text'].count().reset_index()
review_count_per_product.columns = ['Product', 'Review Count']

# Rank products based on review count
top_reviewed_products = review_count_per_product.sort_values(by='Review Count', ascending=False)

# Display the top reviewed products
print("Top Reviewed Amazon Products:")
print(top_reviewed_products.head(20))  # Displaying the top 20 most reviewed products
```

```python
import matplotlib.pyplot as plt
# Rank products based on review count
top_reviewed_products = review_count_per_product.sort_values(by='Review Count', ascending=False).head(20)

# Visualize the top reviewed products
plt.figure(figsize=(10, 6))
sns.barplot(x='Review Count', y='Product', data=top_reviewed_products, palette='viridis')
plt.title('Top Reviewed Amazon Products')
plt.xlabel('Number of Reviews')
plt.ylabel('Product')
plt.show()
```

These top-reviewed products offer valuable insights into consumer preferences and satisfaction levels, showcasing their popularity and positive reception among Amazon shoppers.

**Sentiment Analysis**
Summary: Use the VADER sentiment analysis model to analyze the sentiment of the reviews for the top 20 products. We'll follow these steps:

Task: Load the VADER sentiment analysis model.
Clean and preprocess the review text data.
Apply the VADER model to classify the sentiment of each review.
Aggregate the sentiment scores to determine the overall sentiment for each product.

```python
#install vader sentiment
pip install vaderSentiment
```

```python
# Convert the top reviwed products data to a DataFrame
top_20_products_df = pd.DataFrame(top_reviewed_products)

# Display the DataFrame
print(top_20_products_df.head())
```

```python
# Importing necessary libraries
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Define a function to calculate sentiment scores
def calculate_sentiment(review):
    sentiment_score = analyzer.polarity_scores(review)
    return sentiment_score['compound']

# Rename the column from "name" to "product"
data.rename(columns={'name': 'Product'}, inplace=True)

# Merge top_20_products_df with your_reviews_df on the 'Product' column
merged_df = pd.merge(top_20_products_df, data, on='Product', how='inner')

# Apply sentiment analysis to each review and calculate sentiment score
merged_df['Sentiment Score'] = merged_df['reviews.text'].apply(calculate_sentiment)

# Aggregate sentiment scores to determine overall sentiment for each product
overall_sentiment = merged_df.groupby('Product')['Sentiment Score'].mean().reset_index()

# Determine overall sentiment (positive, neutral, or negative) for each product
overall_sentiment['Sentiment'] = overall_sentiment['Sentiment Score'].apply(lambda x: 'Positive' if x > 0 else ('Neutral' if x == 0 else 'Negative'))

# Display the overall sentiment for each product
print(overall_sentiment)

```

```python
# Plotting the sentiment scores
plt.figure(figsize=(10, 6))
plt.bar(overall_sentiment['Product'], overall_sentiment['Sentiment Score'], color='pink')
plt.xlabel('Product')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Scores for Top 20 Amazon Products')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insights:**

The sentiment analysis reveals overwhelmingly positive feedback for the top 20 Amazon products, with all products garnering positive sentiment scores. Among them, the "Kindle Voyage E-reader" stands out with the highest positive sentiment score. This suggests that consumers are highly satisfied with these products, reflecting positive sentiment towards their features, functionality, and overall user experience. The consistent positive sentiment across diverse product categories underscores Amazon's commitment to delivering high-quality products that meet consumer preferences and expectations.

```python
# Group the data by product name and calculate the count of 5-star ratings for each product
five_star_counts = data[data['reviews.rating'] == 5].groupby('Product')['reviews.rating'].count()

# Sort the products by the count of 5-star ratings in descending order
top_products = five_star_counts.sort_values(ascending=False)

# Display the top products with the most 5-star ratings
print("Top products with the most 5-star ratings:")
print(top_products.head(10))  # Display the top 10 products
```

```python
import matplotlib.pyplot as plt

# Plot the top products with the most 5-star ratings
plt.figure(figsize=(10, 6))
top_products.head(10).plot(kind='bar', color='gold')
plt.title('Top Products with the Most 5-Star Ratings')
plt.xlabel('Product Name')
plt.ylabel('Number of 5-Star Ratings')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

The bar plot illustrates the top products from the dataset based on the number of 5-star ratings they received. The product "AmazonBasics AAA Performance Alkaline Batteries (36 Count)" stands out with the highest number of 5-star ratings, followed by "AmazonBasics AA Performance Alkaline Batteries (48 Count) - Packaging May Vary" and "Fire HD 8 Tablet with Alexa, 8 HD Display, 16 GB, Tangerine - with Special Offers". This visualization provides valuable insights into the most positively rated products, highlighting their popularity among customers.
