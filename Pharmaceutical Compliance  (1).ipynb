## Methodology

**Research Design**

This study adopts a quantitative research design to analyze FDA drug recall data from 2024. The objective is to develop predictive models that enhance compliance documentation processes within pharmaceutical operations. 

**Data Source**

For this study, I utilized the U.S. Food and Drug Administration's (FDA) Enforcement Reports, focusing specifically on drug-related recalls from the year 2024. The FDA publishes weekly Enforcement Reports that detail various enforcement actions, including recalls, market withdrawals, and safety alerts. These reports are publicly accessible and provide comprehensive information pertinent to our analysis. 
(https://www.fda.gov/safety/recalls-market-withdrawals-safety-alerts/enforcement-reports)

**Data Collection**

The dataset was compiled by aggregating weekly Enforcement Reports issued throughout 2024. Each report was meticulously reviewed to extract entries related to drug products. The following key attributes were collected for each recall event:

- **Product Description:** Detailed information about the recalled drug, including its name, dosage form, strength, and packaging details.

- **Reason for Recall:** A concise explanation outlining the specific issue or defect that prompted the recall, such as contamination, mislabeling, or potency deviations.

- **Recalling Firm:** The name of the company responsible for manufacturing or distributing the recalled product.

- **Classification:** The FDA's categorization of the recall based on the severity of health risk posed to the public.

    Class I – Most severe
    The product poses a reasonable probability of causing serious adverse health consequences or death.
    Example: Contaminated sterile injections, mislabeled critical-dose medications.

    Class II – Moderate severity
    The product may cause temporary or medically reversible adverse health effects, or the probability of serious consequences       is    remote.
    Example: Slight potency deviations, non-sterile topical products.

    Class III – Least severe
    The product is unlikely to cause adverse health consequences.
    Example: Minor labeling issues, wrong font size, or packaging defects that don’t affect usage or safety.

By focusing on these attributes, the dataset provides a comprehensive overview of drug recall activities within the specified timeframe, serving as a foundation for subsequent analysis on compliance and documentation optimization in pharmaceutical operations.

## Discussion

1. Timeliness of Recall Reporting
   The analysis revealed a wide disparity between the median and average time taken to report a recall. While the median     reporting time was 33 days, the average was significantly higher (106.7 days), driven by a subset of recalls with extreme delays. In a regulated industry where public safety is paramount, delayed reporting not only increases risk to consumers but can also attract regulatory scrutiny. These delays suggest gaps in internal escalation, documentation, or decision-making workflows, particularly for firms with recurrent Class I or II recalls.

   This pattern signals a need for pharmaceutical companies to re-evaluate internal processes that govern recall reporting timelines. Digital solutions, such as automated document flagging systems, may help reduce bottlenecks and support faster regulatory responses.

2. Severity Classifications and Predictive Modeling
   The classification of recalls into Class I, II, and III provided insight into how firms handle varying degrees of risk. Class I recalls, which pose the greatest threat to patient safety, were significantly associated with issues such as non-sterility and chemical contamination, while Class II recalls dominated the dataset, often tied to CGMP deviations and sterility assurance failures.

This validates the model’s ability to detect latent risk patterns in unstructured data and provides a basis for applying similar approaches in pre-audit documentation review or live monitoring systems.

3. Seasonal Trends and Compliance Planning
   The study found recall activity peaked in January and November, which could be attributed to seasonal audits, post-holiday inspections, or year-end quality reviews. Such predictable trends offer a strategic opportunity for compliance and QA teams to plan resource allocations more effectively.
```python
# Drop columns that aren't useful for modeling
df_model = df2.drop(columns=[
    'Recall Number',
    'Center Classification Date',
    'Recall Initiation Date',
    'Report Date',
    'Voluntary/Mandated',
    'Time to Report'
])

# Drop rows with missing target values for classification
df_model = df_model[df_model['Classification'].notna()]


df_model.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Encode 'Month'
month_map = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}
df_model['Month_Num'] = df_model['Month'].map(month_map)

# Encode 'Recalling Firm'
le_firm = LabelEncoder()
df_model['Firm_Encoded'] = le_firm.fit_transform(df_model['Recalling Firm'].astype(str))

# TF-IDF on text fields
tfidf_reason = TfidfVectorizer(max_features=100)
X_reason = tfidf_reason.fit_transform(df_model['Reason for Recall'].fillna("")).toarray()

tfidf_product = TfidfVectorizer(max_features=100)
X_product = tfidf_product.fit_transform(df_model['Product Description'].fillna("")).toarray()

# Combine features into final X matrix
X = np.hstack([
    X_reason,
    X_product,
    df_model[['Month_Num', 'Firm_Encoded']].fillna(0).values
])
```

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, mean_absolute_error, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# === Targets ===
from sklearn.preprocessing import LabelEncoder
le_class = LabelEncoder()
y_class = le_class.fit_transform(df_model['Classification'].astype(str))
y_days = df_model['Days_to_Report'].astype(float)

# === Train/Test Split ===
X_train, X_test, y_class_train, y_class_test = train_test_split(X, y_class, test_size=0.2, random_state=42)
_, _, y_days_train, y_days_test = train_test_split(X, y_days, test_size=0.2, random_state=42)

# === Classification Model ===
clf = RandomForestClassifier(n_estimators=200, class_weight='balanced', random_state=42)
clf.fit(X_train, y_class_train)
y_class_pred = clf.predict(X_test)

print("📌 Task A - Recall Severity Classification")
print(classification_report(y_class_test, y_class_pred, target_names=le_class.classes_))

# === Regression Model ===
reg = RandomForestRegressor(n_estimators=200, random_state=42)
reg.fit(X_train, y_days_train)
y_days_pred = reg.predict(X_test)

print("\n📌 Task B - Days to Report Regression")
print(f"MAE: {mean_absolute_error(y_days_test, y_days_pred):.2f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_days_test, y_days_pred)):.2f}")
```

 Insight: Class II predictions are strong. Class III predictions show low recall, suggesting potential under-detection of low-severity recalls. Future improvements may include SMOTE or targeted rebalancing.
 
 Insight: Average predictions are reasonably close to actual reporting timelines, but high RMSE indicates a few significant delays. Log transformation will help stabilize outliers.
